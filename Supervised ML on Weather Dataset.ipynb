```python
for i in df.columns:
    if (df[i].dtypes) in [float, int]:
        df[i] = df[i].fillna(df[i].mean())
    elif (df[i].dtypes)==object:
        df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.drop_duplicates(inplace=True)
```

```python
for i in df.columns:
    if df[i].dtypes in [int, float]:
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.ylabel('count')
        plt.show()
```

```python
for i in df.columns:
    if df[i].dtypes in [int, float]:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3-Q1
        lower_fence = Q1 - 1.5* IQR
        upper_fence = Q3 + 1.5*IQR
        df = df[(df[i]>=lower_fence) & (df[i]<=upper_fence)]
```

```python
for i in df.columns:
    if df[i].dtypes in [int, float]:
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.ylabel('count')
        plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in x:
    if x[i].dtypes == object:
        x[i] = le.fit_transform(x[i])
```

```python
x.info()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
for i in x.columns:
    if x[i].dtypes in [float, int]:
        x[i] = sc.fit_transform(x[i].values.reshape(-1, 1))
    
```

# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.8, random_state = 40)
```

```python
rc = RandomForestClassifier()
rc.fit(x_train, y_train)
y_pred = rc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
bagging_model = BaggingClassifier(base_estimator = DecisionTreeClassifier(), n_estimators=10, random_state= 40)
bagging_model.fit(x_train, y_train)
bagging_predictions = bagging_model.predict(x_test)
bagging_accuracy = accuracy_score(y_test, bagging_predictions)
print(bagging_accuracy)
```

```python
boosting_model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=10, random_state = 40)
boosting_model.fit(x_train, y_train)
boosting_predictions = boosting_model.predict(x_test)
boosting_accuracy = accuracy_score(y_test, boosting_predictions)
print(boosting_accuracy)
```

```python
print(bagging_accuracy)
print(boosting_accuracy)
```
